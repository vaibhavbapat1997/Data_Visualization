```python
import pandas as pd #importing the file here
df=pd.read_csv("height_gender.csv") #reading the file
male_height=df.loc[df["Gender"]=="Male"]
```

```python
df
```

```python
import matplotlib.pyplot as plt #this is used to plot a histogram
plt.hist(male_height["Height"], bins=30, facecolor='red', edgecolor='black')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Height of males')
plt.show()
```

```python
#compare the mean and median
male_height.describe()
```

```python
###Check for 1st deviation
one_std=male_height.loc[(male_height["Height"]< 71.889708) & (male_height["Height"]>66.16298400000001)]
print(len(one_std))

###Check for 2nd deviation
two_std=male_height.loc[(male_height["Height"]< 74.75307000000001) & (male_height["Height"]>63.29962200000001)]
print(len(two_std))

###Check for 3rd deviation
three_std=male_height.loc[(male_height["Height"]< 77.616432) & (male_height["Height"]>60.436260000000004)]
print(len(three_std))

```

The distplot represents the univariate distribution of data i.e. data distribution of a variable against the density distribution. 

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
tips=pd.read_csv("tips.csv")

import seaborn as sns #by convention we import  seaborn as sns
import matplotlib.pyplot as plt
#we will see different distribution plots
#try to see how different variables are distributed
sns.histplot(tips["total_bill"],bins=20,color="red",kde=True)
plt.xlabel("Bingo",fontsize=15)
plt.text(25,20,"Writing in sns", size='medium', color='green')
plt.xticks(rotation=90)
```

```python
'''this is using distplot'''
sns.distplot(tips["total_bill"],bins=20,color="red",hist_kws=dict(edgecolor="green", linewidth=2))
```

```python
#simple scatter plots
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="day", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", size="day",sizes=(10, 1000), data=tips)
```

```python
'''doing the regression plot'''
sns.regplot(x="total_bill",y="tip",data=tips)
```

```python
#lets say we want to compare two distributions
sns.jointplot(x="total_bill",y="tip",data=tips)  
```

```python
##also works with hue
sns.jointplot(x="total_bill",y="tip",data=tips, hue="sex")

```

```python
#Add regression line
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

```python
#Replace the scatterplot with a joint histogram using hexagonal bins:
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")

```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
```

```python
#if you are feeling lazy
sns.pairplot(tips) #plots the distribution for all numerical columns
```

```python
#passing the hue argument from a categorical column
sns.pairplot(tips, hue="sex")
```

```python
#changing some colors
sns.pairplot(tips, hue="sex",palette="Set3")
```

```python
#****************************************
#****************************************
"""Plotting the categorical variables"""
#****************************************
#****************************************
#plotting the bar plots
sns.countplot(x="sex",data=tips,hue="smoker")
```

```python
import pandas as pd
titanic=pd.read_csv("Titanic.csv")

import seaborn as sns
sns.pairplot(titanic)

sns.jointplot(x="Age",y="Fare",data=titanic)
sns.histplot(titanic["Fare"])
sns.boxplot(x="Pclass",y="Age",data=titanic)
sns.countplot(x="Sex",data=titanic)

tc=titanic.corr()
sns.heatmap(tc)
sns.relplot(x="Pclass", y="Fare", hue="Sex", data=titanic)
```
